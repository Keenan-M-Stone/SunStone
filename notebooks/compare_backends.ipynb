```python
import json
import numpy as np
import matplotlib.pyplot as plt

def load_monitor_csv(path):
    import csv
    t = []
    v = []
    with open(path) as f:
        reader = csv.DictReader(f)
        for row in reader:
            t.append(float(row.get('t') or 0.0))
            v.append(float(row.get('value') or 0.0))
    return np.array(t), np.array(v)

# Example: t1,v1 = load_monitor_csv('runs/run_meep/outputs/monitors/mon1.csv')
#          t2,v2 = load_monitor_csv('runs/run_opal/outputs/monitors/mon1.csv')
#          plt.plot(t1, v1, label='meep')
#          plt.plot(t2, v2, label='opal')
#          plt.legend(); plt.show()
```
